```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import scipy
%matplotlib inline
plt.viridis()
from matplotlib import animation, rc
from IPython.display import HTML
from brian2.units import *
```

```python
synaptic_data = np.loadtxt("multiTimeline_synaptic.csv", skiprows=1, delimiter=',',
                           dtype={'names': ('date', 'neurogenesis', 'structural_plasticity', 'synaptic_plasticity'),
                          'formats': ('S7', 'i4', 'i4', 'i4')})
deep_data = np.loadtxt("multiTimeline_synaptic_vs_deep_learning.csv", skiprows=1, delimiter=',',
                           dtype={'names': ('date', 'neurogenesis', 'structural_plasticity', 'synaptic_plasticity', 'deep_learning'),
                          'formats': ('S7', 'i4', 'i4', 'i4', 'i4')})
```

```python
f = plt.figure(figsize=(20,10), dpi=1600)
plt.plot(range(synaptic_data.shape[0]), synaptic_data['neurogenesis'], label='Neurogenesis')
plt.plot(range(synaptic_data.shape[0]), synaptic_data['structural_plasticity'], label='Structural plasticity')
plt.plot(range(synaptic_data.shape[0]), synaptic_data['synaptic_plasticity'], label='Synaptic plasticity')
plt.ylabel("Percentage of max")
plt.xlabel("Date")
plt.xticks(np.arange(synaptic_data.shape[0]//4)*4, synaptic_data['date'][np.arange(synaptic_data.shape[0]//4)*4], rotation=45)
plt.tight_layout(.3)
plt.legend(loc='best')
plt.savefig("search_comparison_neurogenesis_sp_synp.png")
```

```python
f = plt.figure(figsize=(20,10), dpi=1600)
plt.plot(range(deep_data.shape[0]), deep_data['neurogenesis'], label='Neurogenesis')
plt.plot(range(deep_data.shape[0]), deep_data['structural_plasticity'], label='Structural plasticity')
plt.plot(range(deep_data.shape[0]), deep_data['synaptic_plasticity'], label='Synaptic plasticity')
plt.plot(range(deep_data.shape[0]), deep_data['deep_learning'], label='Deep learning')
plt.ylabel("Percentage of max")
plt.xlabel("Date")
plt.xticks(np.arange(deep_data.shape[0]//4)*4, deep_data['date'][np.arange(deep_data.shape[0]//4)*4], rotation=45)

plt.tight_layout(.3)
plt.legend(loc='best')
plt.savefig("search_comparison_neurogenesis_sp_synp_deepl.png")
```
