```python
data = np.load("C:/Work/phd/sundry/random_balanced_network_results_113104_24052018.npz")
print data.files
```

```python
sim_time= data['sim_time']
exc_spikes =  data['exc_spikes']
inh_spikes =  data['inh_spikes']
```

```python
def plot_spikes(exc_spikes, inh_spikes, title, simtime=sim_time):
    f, ax1 = plt.subplots(1, 1, figsize=(24, 12), dpi=800)
    ax1.set_xlim((0, simtime))
    ax1.scatter([i[1] for i in exc_spikes], [i[0] for i in exc_spikes],s=1, marker="*")
    
    ax1.scatter([i[1] for i in inh_spikes], [i[0]+500 for i in inh_spikes],s=1, marker="*", c='r')
    ax1.set_xlabel('Time(ms)')
    ax1.set_ylabel('spikes')
    ax1.set_title(title)
    plt.savefig(title + ".png")
    plt.show()
```

```python
plot_spikes(exc_spikes, inh_spikes, "Random Balanced Network")
```
